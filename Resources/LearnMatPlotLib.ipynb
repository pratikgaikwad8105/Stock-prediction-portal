## Subplots and Legends 

```python
days = [1,2,3,4,5]
o_price = [10,20,30,40,50]
p_price = [8, 21,32,44,43]
```

```python
plt.plot(days, o_price, 'b-*', label = 'Original');
plt.plot(days, p_price, 'r-o',  label = 'Predicted');
plt.legend()
plt.title("Plot with Legends");
plt.xlabel('Weekdays');
plt.ylabel('Price');
```

## Subplot

```python
# Original
plt.subplot(3,1,1)
plt.plot(days, o_price, 'b-*', label = 'Original');
plt.legend()
plt.title("Original");
plt.xlabel('Weekdays');
plt.ylabel('Price');

#Predicted
plt.subplot(3,1,2)
plt.plot(days, p_price, 'r-o', label = 'Predicted');
plt.legend()
plt.title("Predicted");
plt.xlabel('Weekdays');
plt.ylabel('Price');

#Difference
price_dif = [o - p for o, p in zip(o_price, p_price)]
plt.subplot(3,1,3)
plt.bar(days, price_dif, color='yellow', label = 'Difference');
plt.legend()
plt.title("Difference");
plt.xlabel('Weekdays');
plt.ylabel('Price');

plt.tight_layout()
```

## Save Figure

```python
plt.plot(days, o_price, 'b-*', label = 'Original');
plt.plot(days, p_price, 'r-o',  label = 'Predicted');
plt.legend()
plt.title("Plot with Legends");
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.savefig('example_prediction', dpi=320)
```
